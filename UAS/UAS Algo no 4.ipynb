**Manajemen Antrian di Taman Hiburan**

Anda adalah seorang insinyur sistem yang bekerja di sebuah taman hiburan besar yang baru saja membuka wahana roller coaster baru yang sangat populer.
Setiap hari, ribuan pengunjung datang untuk menikmati wahana ini, dan manajemen antrian menjadi sangat penting untuk memastikan semua pengunjung mendapatkan pengalaman yang menyenangkan tanpa harus menunggu terlalu lama.

Tantangan Anda:
1.  **Pengelolaan Antrian Prioritas:**
    - Ada tiga jenis tiket yang dijual: Reguler, VIP, dan Super VIP.
    - Pengunjung dengan tiket Super VIP harus mendapatkan prioritas tertinggi, diikuti oleh VIP, dan kemudian Reguler.
    - Buatlah sistem antrian yang memastikan pengunjung diprioritaskan sesuai dengan tiket mereka, namun tetap mempertahankan urutan kedatangan di antara pengunjung dengan jenis tiket yang sama.
2. **Penyeimbangan Waktu Tunggu:**
    - Setiap pengunjung memiliki waktu tunggu maksimum yang mereka bersedia toleransi sebelum meninggalkan antrian. Waktu tunggu maksimum bervariasi tergantung pada jenis tiket mereka.
        - Reguler: 30 menit
        - VIP: 20 menit
        - Super VIP: 10 menit
    - Buatlah sistem yang dapat memantau dan menyesuaikan waktu tunggu untuk memastikan tidak ada pengunjung yang menunggu lebih lama dari waktu maksimum yang ditentukan. Jika waktu tunggu melebihi batas, pengunjung  tersebut harus dikeluarkan dari antrian dan diberikan kompensasi.
    - Pengunjung yang menerima kompensasi akan mendapatkan prioritas padaantrian berikutnya, terlepas dari jenis tiket mereka, dan ditempatkan dalam antrian khusus untuk memastikan mereka mendapatkan akses cepat ke wahana.
3. **Optimalisasi Kapasitas Wahana:**
   - a. Roller coaster hanya dapat menampung 24 orang dalam satu perjalanan.
   - b. Anda harus mengelola antrian untuk memastikan setiap perjalanan roller coaster diisi dengan seoptimal mungkin tanpa melanggar aturan prioritas tiket.
   - c. Jika memungkinkan, setiap perjalanan harus diisi oleh kombinasi pengunjung dari ketiga jenis tiket untuk menjaga kepuasan pelanggan.
   - d. Sistem harus bisa menangani situasi di mana jumlah pengunjung dalam antrian tidak mencukupi untuk mengisi 24 tempat dalam satu perjalanan. Jika antrian tidak penuh, perjalanan tetap dilakukan dengan penumpang yang ada.
4. **Penanganan Ketidakseimbangan Antrian:**
    - Untuk memastikan pengunjung dengan tiket Reguler juga mendapatkan kesempatan untuk naik, tetapkan aturan bahwa setidaknya 20% dari setiap perjalanan harus diisi oleh pengunjung dengan tiket Reguler, jika ada.
    - Jika ada ketidakseimbangan yang signifikan dalam jumlah pengunjung dari masing-masing jenis tiket, sistem harus menyesuaikan alokasi tempat untuk memastikan keadilan.'
5. **Pengunjung yang Datang Bersama-sama:**
    - Jika pengunjung datang dalam kelompok dengan jenis tiket yang berbeda, sistem harus berusaha menjaga kelompok tersebut bersama-sama sejauh mungkin tanpa melanggar aturan prioritas dan kapasitas.
    - Jika perlu, kelompok tersebut dapat dibagi ke perjalanan yang berbeda dengan prioritas tiket yang sama, namun dengan waktu tunggu seminimal mungkin di antara perjalanan tersebut.

```python
from datetime import datetime, timedelta
class antrian:
    def __init__(self):
        self.antrian_reg  = []
        self.antrian_VIP  = []
        self.antrian_SVIP = []
        self.antrian_toleransi = []
        self.total_pemberangkatan = 0
        self.total_pengguna = 0
        self.total_antrian = 0
        self.jam_sekarang = datetime.strptime("08:00:00", "%H:%M:%S")
    
    def pengunjung_baru(self, data_pengunjung):
        self.jam_sekarang = datetime.strptime(data_pengunjung['waktu_kedatangan'], "%H:%M:%S")
        self.total_pengguna += 1
        self.total_antrian += 1
        
        if len(self.antrian_toleransi) != 0 or (len(self.antrian_reg)+len(self.antrian_VIP)+len(self.antrian_SVIP)+len(self.antrian_toleransi) >= 24) or (len(self.antrian_reg)!=0 and len(self.antrian_VIP)!=0 and len(self.antrian_SVIP)!=0):
            if (len(self.antrian_reg)+len(self.antrian_VIP)+len(self.antrian_SVIP)+len(self.antrian_toleransi) >= 24) or self.jam_sekarang >= self.antrian_reg[0]['maks_tunggu'] or self.jam_sekarang >= self.antrian_VIP[0]['maks_tunggu'] or self.jam_sekarang >= self.antrian_SVIP[0]['maks_tunggu']:
                self.meluncur()
                
        print(f'\nwisatawan ke-{self.total_pengguna} hari ini masuk antrian!!!')
        if data_pengunjung['jenis_tiket'] == 'Reguler':
            data_pengunjung.update({'maks_tunggu': (datetime.strptime(data_pengunjung['waktu_kedatangan'], "%H:%M:%S") + timedelta(minutes=30))})
            self.antrian_reg.append((data_pengunjung))
        if data_pengunjung['jenis_tiket'] == 'VIP':
            data_pengunjung.update({'maks_tunggu': (datetime.strptime(data_pengunjung['waktu_kedatangan'], "%H:%M:%S") + timedelta(minutes=20))})
            self.antrian_VIP.append(data_pengunjung)
        if data_pengunjung['jenis_tiket'] == 'Super VIP':
            data_pengunjung.update({'maks_tunggu': (datetime.strptime(data_pengunjung['waktu_kedatangan'], "%H:%M:%S") + timedelta(minutes=10))})
            self.antrian_SVIP.append(data_pengunjung)
        
        self.semua_antrian()        
        self.cek_toleransi()
        
        
    def meluncur(self):
        self.total_pemberangkatan += 1
        berangkat = []
        
        while len(self.antrian_toleransi) != 0:
            berangkat.append(self.antrian_toleransi.pop(0))
            if len(berangkat) ==  24:
                break
        
        sisa_kapasitas_roler = 24 - len(berangkat)
        
        while sisa_kapasitas_roler != 0 or (len(self.antrian_reg)!=0 and len(self.antrian_VIP)!=0 and len(self.antrian_SVIP)!=0):
            
            if len([pengguna for pengguna in berangkat if pengguna['jenis_tiket'] == 'Reguler']) <= 0.2*len(berangkat):
                if len(self.antrian_reg) != 0:
                    berangkat.append(self.antrian_reg.pop(0))
                    
            elemen_0 = []
            if len(self.antrian_reg) > 0:
                elemen_0.append(self.antrian_reg[0])
            if len(self.antrian_VIP) > 0:
                elemen_0.append(self.antrian_VIP[0])
            if len(self.antrian_SVIP) > 0:
                elemen_0.append(self.antrian_SVIP[0])

            if len(elemen_0) == 0:
                break
            
            if min(elemen_0, key=lambda x: x['maks_tunggu']) in self.antrian_reg:
                berangkat.append(self.antrian_reg.pop(0))
            elif min(elemen_0, key=lambda x: x['maks_tunggu']) in self.antrian_VIP:
                berangkat.append(self.antrian_VIP.pop(0))
            elif min(elemen_0, key=lambda x: x['maks_tunggu']) in self.antrian_SVIP:
                berangkat.append(self.antrian_SVIP.pop(0))
            sisa_kapasitas_roler += 1
            
        print(f'\nMELUNCURRR!!!!\njam sekarang:', {self.jam_sekarang.strftime("%H:%M:%S")} if self.jam_sekarang <  min(berangkat, key=lambda x: x['maks_tunggu'])['maks_tunggu'] else (min(berangkat, key=lambda x: x['maks_tunggu'])['maks_tunggu']).strftime("%H:%M:%S") , f'\npemberangkatan ke-{self.total_pemberangkatan}: ', [[i["id"],i['jenis_tiket']] for i in berangkat], '\n')
        self.total_antrian += -(len(berangkat))
        self.semua_antrian()
        
    def cek_toleransi(self):
        def toleransi(isi):
            if len(isi) != 0:
                cek = isi[0]['maks_tunggu']
                while self.jam_sekarang > cek:
                    self.antrian_toleransi.append(isi.pop(0))
                    cek = isi[0]['maks_tunggu']
        toleransi(self.antrian_reg)
        toleransi(self.antrian_VIP)
        toleransi(self.antrian_SVIP)            
        
    def semua_antrian(self):
        a = [[' '.join(['id:', str(i['id']), '|', 'waktu datang:', i['waktu_kedatangan'], '|', 'waktu maks', (i['maks_tunggu']).strftime("%H:%M:%S")])] for i in self.antrian_reg]
        b = [[' '.join(['id:', str(i['id']), '|', 'waktu datang:', i['waktu_kedatangan'], '|', 'waktu maks', (i['maks_tunggu']).strftime("%H:%M:%S")])] for i in self.antrian_VIP]

        c = [[' '.join(['id:', str(i['id']), '|', 'waktu datang:', i['waktu_kedatangan'], '|', 'waktu maks', (i['maks_tunggu']).strftime("%H:%M:%S")])] for i in self.antrian_SVIP]
        d = [[' '.join(['id:', str(i['id']), '|', 'waktu datang:', i['waktu_kedatangan'], '|', 'waktu maks', (i['maks_tunggu']).strftime("%H:%M:%S")])] for i in self.antrian_toleransi]
        print(f'jam sekarang: {self.jam_sekarang.strftime("%H:%M:%S")}\nantrian kategori Reguler: {a}\nantrian kategori VIP: {b}\nantrian kategori Super VIP: {c}\nantrian kategori toleransi: {d}\nbanyak orang mengantri: {self.total_antrian}')
```
